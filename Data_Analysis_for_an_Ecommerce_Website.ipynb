```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Train.csv")
```

```python
print(df.head())
```

```python
print("Dataset shape: ", df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
plt.hist(df["Product_importance"])
plt.title("Histogram of Product Importance")
plt.xlabel("Importance")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x="Discount_offered", y="Product_importance", data=df)
plt.title("Scatterplot of Total Items vs Product Importance")
plt.show()
```

```python
payment_method_group = df.groupby("Cost_of_the_Product")["Prior_purchases"].mean()
print(payment_method_group)
```
